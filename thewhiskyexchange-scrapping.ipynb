```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
base_url = 'https://www.thewhiskyexchange.com/'
```

```python
headers = {
    'user-agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/114.0.0.0 Safari/537.36'
}
```

```python
r = requests.get('https://www.thewhiskyexchange.com/b/40/highland-park-single-malt-scotch-whisky')
soup = BeautifulSoup(r.content, 'lxml')
soup
```

```python
product_list = soup.find_all('li', class_ ='product-grid__item')
product_list
```

```python
product_links = []
for items in product_list:
    for link in items.find_all('a', href = True):
        li = link['href']
        product_links.append(base_url + li)

print(len(product_links))
product_links
```

```python
product_links = []
for x in range(1,5):
    r = requests.get(f'https://www.thewhiskyexchange.com/b/40/highland-park-single-malt-scotch-whisky?pg={x}')
    soup = BeautifulSoup(r.content, 'lxml')
    product_list = soup.find_all('li', class_ ='product-grid__item')

    for items in product_list:
        for link in items.find_all('a', href = True):
            li = link['href']
            product_links.append(base_url + li)

print(len(product_links))
product_links
```

```python
final_list = []
for links in product_links:
    r = requests.get(links, headers = headers)
    soup = BeautifulSoup(r.content, 'lxml')
    
    try:
        name = soup.find('h1', class_ = 'product-main__name').text.strip()
    except:
        'No name'
    try:
        price = soup.find('p', class_='product-action__price').text.strip()
    except:
        'No price'
    try:
        product_meta = soup.find('ul', class_='product-main__meta').text.strip()
    except:
        'No product_meta'
    try:
        ratings = soup.find('span', class_='review-overview__rating star-rating star-rating--40').text.strip()
    except:
        'No ratings'
    
    final = {
        'name':name,
        'price':price,
        'product-meta':product_meta,
        'ratings':ratings,
    }
    
    final_list.append(final)

final_list
    
```

```python
df = pd.DataFrame(final_list)
df.head(15)
```
